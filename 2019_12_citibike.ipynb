```python
import pandas as pd
```

```python
#read csv file
citibike_df = pd.read_csv("201912-citibike-tripdata.csv")
citibike_df.head()
```

```python
#check data types
citibike_df.dtypes
```

```python
# update gender column 
citibike_df["gender"].replace({1: "Male", 2:"Female", 0: "Unspecified"}, inplace=True)
citibike_df.head()
```

```python
# calculate rider age in 2019
citibike_df["Age"] = 2019 - citibike_df["birth year"]
citibike_df.head()
```

```python
citibike_df.to_csv("updated_201912-citibike-tripdata.csv", index = False)
```
